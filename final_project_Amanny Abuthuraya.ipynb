```python
#Amanny Abuthuraya, DATA333 16368, 12/1/2023, Final Project

def menu():
    #to take input from the user
    return int(input("What would you like to do? 1)View top 5 places that people come from \n  2)View top 5 countries that moved to a certain county \n  3)Percentage of people transfering \n 4)Other popular counties \n  5) View most & least month/year transfers"))


def top_5_countries(data):
    #this function will take the countries dataset, count each county and display the top 5 using pandas
    not_usa= data[data["prior_country"]!="USA"]
    count = not_usa["prior_country"].value_counts()
    country = count.head(5)
    df= pd.DataFrame(country)
    display(df)

def top_5_states(data):
    #this function will take the states dataset, count each county and display the top 5 using pandas
    not_wa= data[data["prior_state"].str.lower()!="not applicable"]
    count = not_wa["prior_state"].value_counts()
    state = count.head(5)
    df = pd.DataFrame(state)
    display(df)

def top_5_filter_c(data):
    # this function takes the county the user would like to see, create for it a filter and display the top 5 results using pandas -- this is for the countries dataset
    county=input("Please enter the county you'd like to view data for: ")
    county_filter_data= data[(data["county"].str.lower()== county.lower()) & (data ["prior_country"] != "USA")]
    unique_countries = county_filter_data["prior_country"].value_counts().head(5)
    df = pd.DataFrame({"prior_country": unique_countries.index, "count": unique_countries.values, "county": county.capitalize()})
    df.index +=1
    display(df)

def top_5_filter_s(data):
    # this function takes the county the user would like to see, create for it a filter and display the top 5 results using pandas -- this is for the states dataset
    county=input("Please enter the county you'd like to view data for: ")
    county_filter_data= data[(data["county"].str.lower()== county.lower()) & (data ["prior_state"] != "Not Applicable")]
    unique_states = county_filter_data["prior_state"].value_counts().head(5)
    df = pd.DataFrame({"prior_state": unique_states.index, "count": unique_states.values, "county": county.capitalize()})
    df.index +=1
    display(df)

def percentage_of_county(data):
    #this function takes the county the user wants to view, calculates the percentage of out of country transfers and displaying it
    #to creat a filter on the data based on the county and that they are out of country transfers, then counting them
    county=input("Please enter the county you'd like to see its percentage: ")
    county_filter_data= data[(data["county"].str.lower()== county.lower()) & (data["prior_state"]=="Not Applicable")]
    unique_county= county_filter_data["county"].value_counts()

    #to make sure that the person is out of country transfer
    wa= data[(data["prior_state"]=="Not Applicable")]
    wa_count= len(wa)

     #to get the total of out of country transfers
    percentage= round((unique_county/wa_count*100), 2)
    print("The percentage of out of country transfers: ")
    print(percentage.item())

def percentage_of_state(data):
    #this function takes the county the user wants to view, calculates the percentage of out of state transfers and displaying it
    #to creat a filter on the data based on the county and that they are out of state transfers, then counting them
    county=input("Please enter the county you'd like to see its percentage: ")
    county_filter_data= data[(data["county"].str.lower()== county.lower()) & (data["prior_state"]!="Not Applicable")]
    unique_county= county_filter_data["county"].value_counts()
    
    #to get the total of out of state transfers
    wa= data[data["prior_state"]!="Not Applicable"]
    wa_count= len(wa)

    #calculating the percentage and printing it
    percentage= round((unique_county/wa_count*100), 2)
    print("The percentage of out of state transfers: ")
    print(percentage.item())

def popular_counties_c(data):
    #this function takes a list of counties to use to filter with on the countries dataset
    county=input("Enter the counties you want to filter out seperated by a ',' or a comma: ").lower()
    county_list= [county.strip() for county in county.split(',')]

    #after splitting the list, it applies to it filters to make sure the data used is for out of country transfers
    county_filter_data = data[(~data["county"].str.lower().isin(county_list)) & (data["prior_state"] == "Not Applicable") & (~data["county"].str.lower().eq("unknown"))]
    unique_counties = county_filter_data["county"].value_counts().head(3) #gets top counties after filter

    #to display the data and print it in pandas
    df = pd.DataFrame({"county": unique_counties.index, "count": unique_counties.values})
    df.index +=1
    display(df)

def popular_counties_s(data):
    #this function takes a list of counties to use to filter with on the states dataset
    county=input("Enter the counties you want to filter out seperated by a ',' or a comma: ").lower()
    county_list= [county.strip() for county in county.split(',')]

    #after splitting the list, it applies to it filters to make sure the data used is for out of state transfers
    county_filter_data = data[(~data["county"].str.lower().isin(county_list)) & (data["prior_state"] != "Not Applicable") & (~data["county"].str.lower().eq("unknown"))]
    unique_counties = county_filter_data["county"].value_counts().head(3) #gets top counties after filter

     #to display the data and print it in pandas
    df = pd.DataFrame({"county": unique_counties.index, "count": unique_counties.values})
    df.index +=1
    display(df)

def most_least_transfers_c(data):
#this function groups data based on year and month for out of country transfers, then gets the max and min and displays it to the user
    out_of_country= data[data["prior_state"].str.lower().eq("not applicable")]
    counts = out_of_country.groupby(["year", "month"]).size().reset_index(name="transfer_count")

    most_transfers = counts[counts["transfer_count"] == counts["transfer_count"].max()]
    least_transfers = counts[counts["transfer_count"] == counts["transfer_count"].min()]

    print("Month and year with the most transfers:")
    print(most_transfers)
    print()
    print("Month and year with the least transfers:")
    print(least_transfers)

def most_least_transfers_s(data):
    #this function groups data based on year and month for out of state transfers, then gets the max and min and displays it to the user
    out_of_state= data[data["prior_state"].str.lower() != "not applicable"]
    counts = out_of_state.groupby(["year", "month"]).size().reset_index(name="transfer_count")

    most_transfers = counts[counts["transfer_count"] == counts["transfer_count"].max()]
    least_transfers = counts[counts["transfer_count"] == counts["transfer_count"].min()]

    print("Month and year with the most transfers:")
    print(most_transfers)
    print()
    print("Month and year with the least transfers:")
    print(least_transfers)

def classify_counties(data):
  # Creating a column called 'county' to represent counties in the license dataset
    count = data['county'].value_counts().reset_index(name='transfer_count')
    count.columns = ['county', 'transfer_count']

    # Creating classification levels by using bins
    bins = [-float('inf'), 500, 1000, 1500, 2000, float('inf')]
    labels = ['Level 1', 'Level 2', 'Level 3', 'Level 4', 'Level 5']

    # Create a new column for classification
    count['budget_level'] = pd.cut(count['transfer_count'], bins=bins, labels=labels, right=False)

    # Display the result
    display(count[['county', 'transfer_count', 'budget_level']])

def conference_months(data):
# this functions group data by month and year, creates for it a column called transfer count by counting each month and displaying the smallest 2
    counts = data.groupby(["year", "month"]).size().reset_index(name="transfer_count")
    least_transfers = counts.nsmallest(2, "transfer_count")

    print("Month and year with the least transfers:")
    print(least_transfers)

def top_10_countries(data):
    #this function will take the license dataset, group the counts by both country and county and display the top 10 using pandas

    #filter out entries with prior_country as "USA"
    not_usa = data[data["prior_country"] != "USA"]

    #group by both "prior_country" and "county" and get the count for each group
    country_county_counts = not_usa.groupby(["prior_country", "county"]).size().reset_index(name='Count')

    #get the top 10 groups based on count
    top_10_countries_counties = country_county_counts.sort_values(by='Count', ascending=False).head(10)
    display(top_10_countries_counties)



import pandas as pd

#giving the program the data source on my PC. Source link will have to be changed before testing
license_data=pd.read_csv("C:\\Users\\amann\\Desktop\\Bellevue College\\Fall 2023\\DATA 333\\Final_Project\\license_data.csv")

#creating two data sets, one for counties and one for states
country_data= license_data[["county","prior_country", "prior_state", "year", "month"]]
state_data= license_data[["county","prior_country", "prior_state", "year", "month"]]
print("Hello, welcome to my final project for Data333. Hope you enjoy it!")
# the try and catch will be used to catch any errors
try:
    y = "yes"
    while y.lower() == "yes":

        # another try statement to get any wrong input
        try:

            #to display what the user would like to do
            data_choice1 = int(input("What would you like to do? 1) Calculate data   2) See business decisions 3) Exit: "))

            if data_choice1 == 1:
                data_choice2 = input("What dataset would you like to work with? Country or State: ")
                if data_choice2.lower() == "country":
                    choice = menu()
                    if choice == 1:
                        top_5_countries(country_data)
                    elif choice == 2:
                        top_5_filter_c(country_data)
                    elif choice == 3:
                        percentage_of_county(country_data)
                    elif choice == 4:
                        popular_counties_c(country_data)
                    elif choice == 5:
                        most_least_transfers_c(country_data)
                    else:
                        print("Please enter a valid input")

                elif data_choice2.lower() == "state":
                    choice = menu()
                    if choice == 1:
                        top_5_states(state_data)
                    elif choice == 2:
                        top_5_filter_s(state_data)
                    elif choice == 3:
                        percentage_of_state(state_data)
                    elif choice == 4:
                        popular_counties_s(state_data)
                    elif choice == 5:
                        most_least_transfers_s(state_data)
                    else:
                        print("Please enter a valid input")


                else:
                    print("Please enter a valid input")

                y = input("Would you like to continue? Yes or No: ")
                if y.lower() == "no":
                    print("You've closed the menu")
                elif y.lower() != "yes":
                    print("Please enter a valid input")

            elif data_choice1 == 2:

                #another menu for the business decisions part of the project
                decision = int(input("Please choose what you'd like to decide upon: 1) Budget reallocation 2) Best 2 months for conferences 3) Brochure decision: "))
                if 1 <= decision <= 3:
                    if decision == 1:
                        classify_counties(license_data)
                    elif decision == 2:
                        conference_months(license_data)
                    elif decision == 3:
                        top_10_countries(license_data)
                    else:
                        print("Please enter a valid input")
                else:
                    print("Please enter a valid input")

            elif data_choice1 == 3:
                break

            else:
                print("Please enter a valid number from the menu")

        except ValueError:
            print("Invalid input. Please enter an option from the menu.")
            continue

except Exception as e:
    print(f"An error occurred: {e}")
else:
    print("Thank you for trying my project!")
```
